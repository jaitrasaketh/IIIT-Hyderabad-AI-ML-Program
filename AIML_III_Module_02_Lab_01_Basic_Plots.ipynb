##  Self implementation of the lab and Notes

```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
url="https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response=requests.get(url)
```

```python
from io import StringIO
memloc_data=StringIO(response.text)
data=pd.read_csv(memloc_data)

data.head()
```

```python
data.replace("?",np.nan,inplace=True)
data.dropna(inplace=True)
len(data)
```

```python
X=data.drop("price",axis=1)
y=pd.to_numeric(data['price'])
```

scatter plot : car companies against their price.

```python
plt.figure(figsize=(24,7))
plt.scatter(X["make"],y)
plt.xlabel("Manufacturer")
plt.ylabel("Price")
plt.title("Manufacturer vs Price- Scatter Plot")
plt.show()
```

Box Plot Visulatisation using seaborn library

```python
sns.set(rc={"figure.figsize":(24,7)})
sns.boxplot(x=X["make"],y=y,palette="Set1").set_title("Manufacturer vs Price - Box Plot")
```

```python
sns.violinplot(x=X["make"],y=y, palette="Set3").set_title('Car maker vs Price - Violin Plot')
```

```python
sns.swarmplot(x=X["make"],y=y).set_title('Car maker vs Price Swarm Plot')
```

Difference between swarm plot and scatter plot:


Scatter plot: Both X and Y axes are continuous variables. The distribution is about relationship between X and Y

Swarm plot: X is categorical variable, Y is continuous variable, the distribution is about how Y is distributed among the categories(X)

```python
sns.boxplot(x=X["make"],y=pd.to_numeric(X["horsepower"]), palette="Set3").set_title('Car maker vs Horsepower Box Plot')
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.scatterplot(x=pd.to_numeric(X["horsepower"]), y = y, color = "red")
```

```python
sns.jointplot(x=pd.to_numeric(X['horsepower']),y=y,kind="reg",color='black')
```

Scatter plot shows distribution between 2 attributes, whereas the margins show histograms that show the distribution variation of that single attribute in the overall distribution

2) Do you think maximum/minimum values are a clear indication of which body-type is costliest, or does the interpretation improve by also taking the interquartile range into consideration?


##### 2. Maximum/Minimum Values vs. Interquartile Range (IQR)

- Maximum and minimum values give a range, but they can be influenced by outliers, which might not represent the typical pricing.
- The IQR (the range between the first quartile (Q1) and third quartile (Q3)) provides a better understanding of the central tendency and spread of the data, showing where the bulk of the prices lie.
- For example, the convertible has the highest IQR, indicating a consistently higher price range compared to other body types.
- Therefore, considering the IQR along with the maximum and minimum values improves the interpretation of which body-type is the costliest.


3) Do you think the body-style of a car can be a good predictor for its price?


##### 3. Body-Style as a Predictor for Price

- The box plot shows distinct differences in price ranges for different body styles, suggesting a relationship between body style and price.
- However, body style alone might not be sufficient as a sole predictor. Other factors like brand, engine type, features, and market conditions could also play significant roles.
- Hence, while body-style can be a good predictor, it should ideally be used alongside other variables to more accurately predict car prices.

```python
## we will be comparing rest of the parameters/columns present in the data with respect to precence or absece of heart disease
data['target'] = data.target.replace({1: "Disease", 0: "No_disease"})
data['sex'] = data.sex.replace({1: "Male", 0: "Female"})
data['cp'] = data.cp.replace({1: "typical_angina",
                          2: "atypical_angina",
                          3:"non-anginal pain",
                          4: "asymtomatic"})
data['exang'] = data.exang.replace({1: "Yes", 0: "No"})
data['slope'] = data.cp.replace({1: "upsloping",
                          2: "flat",
                          3:"downsloping"})
data['thal'] = data.thal.replace({1: "fixed_defect", 2: "reversable_defect", 3:"normal"})
```

```python
pca = PCA()
pca.fit(numeric_data)

# Plot the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs. Number of Principal Components')
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(numeric_data)
pca_data = pca.transform(numeric_data)

# Create a DataFrame with the principal components and target labels
pca_df = pd.DataFrame({
    "pca_1": pca_data[:, 0],
    "pca_2": pca_data[:, 1],
    "target": data["target"]
})

# Visualize the PCA results with a scatter plot
sns.scatterplot(x="pca_1", y="pca_2", hue="target", data=pca_df)
plt.title("PCA Visualization of Heart Disease Data")
plt.show()
```

```python
from sklearn.manifold import TSNE

# Initialize and fit the TSNE model
tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(numeric_data)

# Create a DataFrame with the TSNE components and target labels
tsne_df = pd.DataFrame({
    "tsne_1": tsne_data[:, 0],
    "tsne_2": tsne_data[:, 1],
    "target": data["target"]
})

# Visualize the TSNE results with a scatter plot
sns.scatterplot(x="tsne_1", y="tsne_2", hue="target", data=tsne_df)
plt.title("TSNE Visualization of Heart Disease Data")
plt.show()
```

cleaning and filling the missing values in the data

```python
data['Caffeine (mg)'] = data['Caffeine (mg)'].replace('Varies', np.NaN).replace('varies', np.NaN)
data['Caffeine (mg)'] = data['Caffeine (mg)'].astype(np.float64)
data['Caffeine (mg)'] = data['Caffeine (mg)'].fillna(data['Caffeine (mg)'].mean())
```

```python
data['Total Fat (g)'].unique()
```

```python
data['Total Fat (g)'] = data['Total Fat (g)'].replace('3 2', '3.2')
```

```python
data.info()
```

```python
# Extract columns with int and float types
numeric_columns = data.select_dtypes(include=["int64", "float64"]).columns

# Print the numeric columns
print(numeric_columns)
```

We will be analysing the dataset using the fact that weather the drink comes under the category tea or not

```python
data['Beverage_category'].unique()
```

```python
data['Tea'] = data['Beverage_category'].apply(lambda x: 1 if x == 'Tazo® Tea Drinks' else 0)
data = data.drop('Beverage_category', axis=1)
```

```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python

data = data.replace({True: 1, False: 0})
```

```python
data = data.applymap(lambda x: np.float64(str(x).replace('%', '')))
```

```python
## to do

# Create a pie chart of the 'Tea' column also write your observation form the plot
plt.pie(data['Tea'].value_counts(), labels=data['Tea'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Tea Consumption')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Load your data
# df = pd.read_csv('your_data.csv')

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Perform PCA
pca = PCA()
pca.fit(scaled_data)

# Plot the explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# Determine the optimal number of principal components
explained_variance = pca.explained_variance_ratio_.cumsum()

# Check if we can reach the 95% variance threshold
if explained_variance[-1] >= 0.95:
    optimal_components = next(i for i, total_variance in enumerate(explained_variance) if total_variance >= 0.95) + 1
else:
    optimal_components = len(explained_variance)

print(f"The optimal number of principal components is {optimal_components}")
```

```python
## to do
# plot the first 2 components of tsne, whats you observation from the plot?
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
tsne = TSNE(n_components=2, random_state=42)
tsne_components = tsne.fit_transform(scaled_data)

plt.figure(figsize=(10, 8))
plt.scatter(tsne_components[:, 0], tsne_components[:, 1], c='blue', edgecolor='k', s=50)
plt.title('t-SNE Visualization of the First 2 Components')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.grid(True)
plt.show()
```

### Observations from the t-SNE Plot

1. **Clusters**: The plot shows several groupings of points, indicating that there are some inherent clusters in the data. These clusters suggest that the data has natural groupings based on the features.

2. **Separation**: There are a few distinct clusters that are well separated from others. This indicates that there are some groups in the data that are quite different from the rest.

3. **Overlap**: There are regions where points are more densely packed and overlap with each other. This indicates that there are areas in the feature space where the data points are quite similar to each other.

4. **Outliers**: There are some points that are far from the main clusters, suggesting the presence of outliers or unique data points that do not conform to the general pattern of the data.

5. **Distribution**: The points are spread out across the plot, indicating a diverse range of feature values in the data. However, certain areas have higher densities, suggesting that some feature combinations are more common than others.

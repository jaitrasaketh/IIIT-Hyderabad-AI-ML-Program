

>  Multi_dot method more efficient than Serialized dot products:




*   CPU times: user 2.56 ms by the multi_dot method

> Reason




 The serialized dot method takes more time than multi_dot method ,as outer dot product gets executed iff inner dot product is executed. Whereas the multi_dot method  computes all the corresponding coefficients at once and finally sums them all thereby saving time

# Singular Value Decomposition

SVD can be used to reduce the amount of data needed to represent an image. By decomposing the image into its singular values and vectors, it's possible to approximate the original image with fewer components, retaining most of the important features while significantly reducing the file size.

```python
from skimage import data
#skimage is a library with 18 genreal purpose images already https://scikit-image.org/docs/stable/auto_examples/data/plot_general.html
image=data.rocket()
print(image.dtype,image.min(),image.max(),image.shape)
##image is a numpy array with pixels ranging from 0 to 255
plt.imshow(image)
plt.show()
```

```python
img=image/255
##normalising so that pixels values come between 0 to 1

img=np.transpose(img,axes=(2,0,1))
# axes are reordered, to change layout of image data to meet expected input format of internal libraries which is colorchannel,height,width



```


   

*   0- height
*   1- width
*   2- color channels

now we changed the layout to be 2->1->0 from the default 0->1->2









```python
U,S,V=np.linalg.svd(img)  ## decomposes image to 3 components
print(U.shape,S.shape,V.shape)
```

```python
Sigma=np.zeros((3,427,640)) ## create a 0 matrix of shape that is compatible to multiply with U and V
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:],S[i,:])   ##fill the diagonal of sigma matrix with the values of ith row of S
print(Sigma.shape)
## each 427x640 matrix corresponds to one color channel with singular values on the diagonal
```

```python
reconst=U@Sigma@V
reconst=np.transpose(reconst,axes=(1,2,0))
```

Now we reorder it back to original position


*   0- color channel
*   1- height
*   2-width

therefore we change from 0->1->2 to 1->2->0






```python
plt.imshow(reconst)
## now reconst will have shape of (427,640,3) which is standard layout for displaying images
plt.show()
```



> Now image reconstruction will be done by reducing the dimensions



```python

#k=50
#k=20
k=15
reconst=U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst=np.transpose(reconst,axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```



> We can toggle the values of k, to see how many datapoints are actually required to see the image clearly
